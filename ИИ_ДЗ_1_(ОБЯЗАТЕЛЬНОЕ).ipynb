```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
from multiprocessing import Pool
import matplotlib.pyplot as plt


data = pd.read_csv('temperature_data.csv')

data['rolling_avg'] = data.groupby('city')['temperature'].rolling(window=30, min_periods=1).mean().reset_index(level=0, drop=True)
# Рассчитаем среднее и стандартное отклонение для каждого сезона и города
season_stats = data.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()

# Функция для выявления аномалий
def detect_anomalies(group):
    # Получаем среднее и стандартное отклонение для города и сезона
    city = group['city'].iloc[0]
    season = group['season'].iloc[0]

    # Извлекаем среднее и стандартное отклонение из season_stats
    season_stat = season_stats[(season_stats['city'] == city) & (season_stats['season'] == season)].iloc[0]
    mean_temp = season_stat['mean']
    std_temp = season_stat['std']

    # Выявляем аномалии: температура выходит за пределы (mean ± 2 * std)
    group['anomaly'] = group['temperature'].apply(lambda x: abs(x - mean_temp) > 2 * std_temp)
    return group

# Выявление аномалий для каждого города и сезона
anomalies = data.groupby(['city', 'season']).apply(detect_anomalies)

anomalies.to_csv('temperature_anomalies.csv', index=False)

# Визуализация аномалий
def plot_anomalies(group):
    plt.figure(figsize=(10, 5))
    plt.plot(group['timestamp'], group['temperature'], label='Temperature')
    plt.scatter(group[group['anomaly']]['timestamp'], group[group['anomaly']]['temperature'], color='red', label='Anomaly')
    plt.title(f"Temperature anomalies for {group['city'].iloc[0]}")
    plt.xlabel('Date')
    plt.ylabel('Temperature')
    plt.legend()
    plt.show()

# Визуализация для каждого города
for city in anomalies['city'].unique():
    city_data = anomalies[anomalies['city'] == city]
    plot_anomalies(city_data)

# Распараллеливание анализа аномалий
def parallel_anomaly_detection(cities):
    with Pool() as pool:
        pool.map(detect_anomalies, [data[data['city'] == city] for city in cities])

# Замер времени выполнения
import time

#start_time = time.time()
#parallel_anomaly_detection(data['city'].unique())
#print(f"Parallel processing time: {time.time() - start_time} seconds")

start_time = time.time()
for city in data['city'].unique():
    detect_anomalies(data[data['city'] == city])
print(f"Single processing time: {time.time() - start_time} seconds")
```
